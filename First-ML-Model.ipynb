```python
# Import the pandas lib to deal with tabular data [Excel, Google sheets, SQL file]
import pandas as pd

# the path to the acutal tabular data.
DATA_FILE_PATH = "./melb_data.csv"

# Read the data into what it's called [DataFrame - imagine it like a table that consists of rows and columns]
home_data = pd.read_csv(DATA_FILE_PATH)
```

```python
# get the first five rows
home_data.head()
```

```python
# drop missing values, cuz I noticed that some data are NaN
home_data = home_data.dropna(axis=0)

# get the first five rows again to get some insigts that there's no missing data
home_data.head()
```

```python
# Generate some descriptive statistics about the data like [count, mean, standard deviation, min, max] values.
home_data.describe()
```

```python
# print the list of columns in the dataset to find the name of the prediction target
home_data.columns
```

```python
# Selecting the predection target, by convention we name it (y)
y = home_data.Price
```

```python
# choosing the features that we gonna extract patterns from it to map to the predection target, by convention this data called (X)
selected_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']

# Select the df data of these features
X = home_data[selected_features]
```

```python
# Review the data that we selected to predecit the price.
X.describe()
```

```python
X.head()
```

```python
# Building model
"""
    This step consist of 4 steps:
        1- Define the model.
        2- Fit the model (capture patterns from provided data).
        3- Predict.
        4- Evaluate (Determine how accurate the model's predictions are.)
"""

```

```python
# import the scikit-learn lib [we use it like sklearn], but here we use a Supervised ML Model called DecisionTreeRegressor, Regressor cuz we deal with some regression data.
from sklearn.tree import DecisionTreeRegressor
```

```python
# Define our model, we use random_state argument with a number to ensure the same result each run for the model.
home_data_model = DecisionTreeRegressor(random_state=1)
```

```python
# Fit (train) the model, to extract the patterns that maps between input(features, X) and output(predection value, y)
home_data_model.fit(X, y)
```

```python
# Predict the price for the first 5 rows of the data, in general we need to make predections for new houses, but for now, we will stick with that.
# that's why the prediction of the price gonna be the same, cuz the model is already know about the following 5 houses
print("Making predictions for the following 5 houses:")
print(X.head())

print("\nThe actual home data price for the first 5 rows:")
print(home_data.Price.head())

print("\nThe predictions are:")
print(home_data_model.predict(X.head()))
```
